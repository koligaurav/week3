```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cartoon = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/cartoon.bmp")
cartoon_y = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/cartoon_y.jpg")
girl = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/girl.png") #grayscale image
girl_y = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/girl_y.jpg")
fingerprint = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/fingerprint.jpg")
fingerprint_y = cv2.imread("/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/fingerprint_y.jpg")
```

```python
fig, axs = plt.subplots(3,2,figsize=(8,12))
axs[0,0].imshow(cv2.cvtColor(cartoon,cv2.COLOR_BGR2RGB))
axs[0,1].imshow(cartoon_y)
axs[0,0].axis("off")
axs[0,1].axis("off")
axs[0,0].set_title("Raw")
axs[0,1].set_title("Masked")
axs[1,0].imshow(girl)
axs[1,1].imshow(girl_y)
axs[1,0].axis("off")
axs[1,1].axis("off")
axs[1,0].set_title("Raw")
axs[1,1].set_title("Edges")
axs[2,0].imshow(cv2.cvtColor(fingerprint,cv2.COLOR_BGR2RGB))
axs[2,1].imshow(fingerprint_y)
axs[2,0].axis("off")
axs[2,1].axis("off")
axs[2,0].set_title("Raw")
axs[2,1].set_title("Binary")
plt.show()
```

```python
bart_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

# HSV conversion for color masking
hsv_image = cv2.cvtColor(bart_rgb, cv2.COLOR_RGB2HSV)

# limits
lower_green = np.array([40, 40, 40])
upper_green = np.array([80, 255, 255])

# masking
mask = cv2.inRange(hsv_image, lower_green, upper_green)

cartoon_y_mentee = cv2.bitwise_and(bart_rgb, bart_rgb, mask=mask) #RGB format

plt.imshow(cartoon_y_mentee)
plt.title('Masked')
plt.axis('off')
plt.show()
```

```python
# changing to grayscale
gray_image = cv2.cvtColor(girl, cv2.COLOR_BGR2GRAY)

# GaussianBlur to reduce noise
blurred_image = cv2.medianBlur(gray_image, 5)

# edge detection using canny
edges = cv2.Canny(blurred_image, 150, 100)
girl_y_mentee = edges

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.axis('off')
```

```python
path = "/content/drive/MyDrive/LS2024/ML/WEEK3/OPENCV/images/fingerprint.jpg"
fingerprint = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

_, fingerprint_y_mentee = cv2.threshold(fingerprint, 100, 255, cv2.THRESH_BINARY_INV)

plt.imshow(fingerprint_y_mentee, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```

```python
fig, axs1 = plt.subplots(3,2,figsize=(8,12))
axs1[0,0].imshow(cartoon_y_mentee)
axs1[0,1].imshow(cartoon_y)
axs1[0,0].axis("off")
axs1[0,1].axis("off")
axs1[0,0].set_title("Masked_Mentee")
axs1[0,1].set_title("Masked")
axs1[1,0].imshow(girl_y_mentee,cmap="gray")
axs1[1,1].imshow(girl_y)
axs1[1,0].axis("off")
axs1[1,1].axis("off")
axs1[1,0].set_title("Edges_Mentee")
axs1[1,1].set_title("Edges")
axs1[2,0].imshow(fingerprint_y_mentee,cmap="gray")
axs1[2,1].imshow(fingerprint_y)
axs1[2,0].axis("off")
axs1[2,1].axis("off")
axs1[2,0].set_title("Binary_Mentee")
axs1[2,1].set_title("Binary")
plt.show()
```
